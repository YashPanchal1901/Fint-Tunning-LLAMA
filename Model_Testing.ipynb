```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFacePipeline, ChatHuggingFace
```

```python
pdf = PyPDFLoader("/content/Alex Net.pdf")
data = pdf.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = splitter.split_documents(data)
```

```python
texts
```

```python
model_name = "BAAI/bge-large-en"
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': False}
embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
load_vectorstore = Chroma(persist_directory="db", embedding_function=embeddings)
retriever = load_vectorstore.as_retriever(search_kwargs={"k":1})
```

```python
repo_id = "YashPanchal1901/my-gguf-model"

llm = HuggingFacePipeline.from_model_id(
    model_id=repo_id,
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens=512,
        do_sample=False,
        repetition_penalty=1.03,
    ),
)

chat_model = ChatHuggingFace(llm=llm)
```

```python
prompt_template = """Use the following pieces of information to answer the user's question.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}
Question: {question}

Only return the helpful answer below and nothing else.
Helpful answer:
"""
prompt = PromptTemplate(template=prompt_template, input_variables=['context', 'question'])
```

```python
from langchain.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(
    llm=chat_model,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents = True,
    chain_type_kwargs= {"prompt": prompt},
    verbose=True
)
```

```python
response = qa.invoke('what is alex net model')
```

```python
response['result']
```

```python
import re
def get_assistant_answer(text):
    # Try with <|eot_id|> first
    match = re.search(r"<\|start_header_id\|>assistant<\|end_header_id\|>\s*(.*?)(?:<\|eot_id\|>|$)", text, re.DOTALL)
    return match.group(1).strip() if match else None
```

```python
from IPython.display import display, Markdown
display(Markdown(get_assistant_answer(response['result'])))
```
